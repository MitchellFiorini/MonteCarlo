# Monte Carlo Integration: Using Monte Carlo Method to determine pi

We will be using the Monte Carlo method to determine the value of pi. 
First, we will need to be able to create random points on a 2D plot.
To do this, we will first create a random number generator.
The range of the random numbers can be any length, but for ease we will choose the range to be between 0 and 1.

```python
import random

#function to generate a random point
def generatePoint():
    point = []
    x = random.uniform(0,1)
    point.append(x)
    y = random.uniform(0,1)
    point.append(y)
    return point
```

```python
#test our random point generator
print(generatePoint())
```

Next we will have to decide which of the random points fall within the circle or outside the circle.
Our circle will have a radius of 1 in this example. 
We will use the distance formula to calculate the distance between the origin of our plot and 1.


```python
#function to calculate distance from origin
import math
def calcDistance(alist):
    temp = math.sqrt((((alist[0])**2) + ((alist[1])**2)))
    #print(temp)
    return temp
```

Now that we have our point generator and can determine which points are inside or outside our circle, we can start plotting.
We will start with only 100 points to plot.

```python
import matplotlib.pyplot as plt
def circle(x):
    count = 0
    #incircle = 0
    while count < x:
        newlist = generatePoint()
        if calcDistance(newlist) < 1.0:
            #incircle+=1
            plt.scatter(newlist[0],newlist[1], color='green')
        else:
            plt.scatter(newlist[0],newlist[1], color='red')
        count+=1
    #print(incircle/count * 4)
    plt.xlim(0,1.0)
    plt.ylim(0,1.0)
    plt.show()
```

```python
circle(100)
```

It is hard to tell if our plot is actually showing a circle or not so we will increase the number of points to be able to see our circle.

```python
circle(1000)
```

```python
circle(5000)
```

Notice our plot is only be positive and shows only 1/4 of the circle.
To account for this we will multiply our output by 4 to get the area of a full circle.
But what is our output?


Our output will be the number of points inside the circle (green) divided by the number of total points (count).
For the sake of speed, we will create a function that does not plot the points, but will output the number of points 
in our 1/4 circle divided by the total number of points * 4.
Note: you need to multiply your output by the area of your plot. Since the area of our plot is 1, multiplying our output will not change the result.

```python
def determinePi(x):
    count = 0
    incircle = 0
    while count < x:
        newlist = generatePoint()
        if calcDistance(newlist) < 1.0:
            incircle+=1
        count+=1
    print(incircle/count * 4)
```

Now let's use our new function with increasing inputs.

```python
determinePi(100)
determinePi(1000)
determinePi(10000)
determinePi(100000)
determinePi(1000000)
determinePi(10000000)
```

As we can see, the more "points" we plot in our circle, the closer we come to the value of pi.
